## Data Cleaning

- cleaning the raw_content column
- splitting the coordinates column into latitude and longitude columns
- getting rid of words that start with '@'


```python
df1 = df.copy()

df1['raw_content'] = df1['raw_content'].str.replace('\n\n', ' ')
df1['raw_content'] = df1['raw_content'].str.replace(';\n', ' ')
df1['raw_content'] = df1['raw_content'].str.replace('\n', ' ')

df1['coordinates'] = df1['coordinates'].astype(str)
df1['coordinates'] = df1['coordinates'].str.extract('\((.*?)\)')
df1['coordinates'] = df1['coordinates'].str.replace('longitude=', '')
df1['coordinates'] = df1['coordinates'].str.replace('latitude=', '')

df1['raw_content'] = df1['raw_content'].str.replace(r'@\w+', '@user')

df1[['longitude', 'latitude']] = df1['coordinates'].str.split(',', expand=True)

df1['latitude'] = pd.to_numeric(df1['latitude'])
df1['longitude'] = pd.to_numeric(df1['longitude'])

df1 = df1.drop(columns='coordinates')

display(df1)
```
